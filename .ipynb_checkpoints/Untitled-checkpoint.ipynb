```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
```

```python
import pandas as pd
data=pd.read_csv("js.CSV",encoding='ansi')
data
```

```python
plt.rcParams[
     'font.sans-serif']=[
     'SimHei']
map_city=nx.Graph()
for i in range(len(data)):
    map_city.add_node(data.city[i],x=data.x[i],y=data.y[i])
pos=dict(zip(data.city,zip(data.x,data.y)))
nx.draw(map_city,pos=pos,node_color='y',node_size=50,with_labels=True)
plt.show()
```

```python
from geopy import distance
edge_matrix=np.zeros((len(data),len(data)))#初始化邻接矩阵全为0
for i in range(len(data)-1):
    for j in range(i+1,len(data)):
        citys_distance=distance.distance((pos[data.city[i]][1],pos[data.city[i]][0]),(pos[data.city[j]][1],pos[data.city[j]][0])).km
        edge_matrix[i][j]=citys_distance
        edge_matrix[j][i]=citys_distance
        map_city.add_edge(data.city[i],data.city[j],weight=citys_distance)#添加边以及边的距离weight
nx.draw(map_city,pos=pos,node_color='y',edge_color='y',node_size=50,with_labels=True)
plt.savefig("plot.png", dpi=800)
plt.show()
```

##算法一：prim算法

```python
import random as rd 
import math

def prim(matrix):#输入为带权邻接矩阵
    node_start=rd.choice(range(len(data)))#随机选一点开始
    U={node_start}
    V=set(range(len(data)))
    edge_prim=[]#初始化最小生成树边集为空
    V=V-U
    for i in range(len(data)-1):
        temp=matrix[np.ix_(list(U),list(V))]
        min=np.argmin(temp)#找当前以U为始发点的V为终结点的最小权重边
        x,y=divmod(min,temp.shape[1])#返回最小权重边的端点位置
        edge_prim.append((list(U)[x],list(V)[y]))
        U.add(list(V)[y])
        V.remove(list(V)[y])
    return edge_prim
        
```

```python
edge_prim=prim(edge_matrix)
print(edge_prim)
```

```python
from IPython import display
map_temp=nx.Graph()
for i in range(len(data)):
    map_temp.add_node(data.city[i],x=data.x[i],y=data.y[i])
temp=dict()
for i,j in edge_prim:
    display.clear_output(wait=True)
    citys_distance=distance.distance((pos[data.city[i]][1],pos[data.city[i]][0]),(pos[data.city[j]][1],pos[data.city[j]][0])).km
    edge_matrix[i][j]=citys_distance
    map_temp.add_edge(data.city[i],data.city[j],weight=citys_distance)
    temp[data.city[i],data.city[j]]=int(citys_distance)
    nx.draw(map_temp,pos=pos,node_color='y',edge_color='y',node_size=50,with_labels=True)
    nx.draw_networkx_edge_labels(map_temp,pos=pos,edge_labels=temp)
    plt.show()
    plt.pause(0.4)
```

```python
import operator
edge={}
def sort_edges(matrix):
    for i in range(len(data)-1):
        for j in range(i+1,len(data)):
           edge[(i,j)]=matrix[i][j]
    sort_edge=sorted(edge.items(),key=operator.itemgetter(1))
    return sort_edge
            
```

获得可达矩阵

```python
def get_reach_matrix(matrix):#入口参数为邻接矩阵
    import numpy as np
    A = np.mat(matrix)
    I = np.identity(len(A))
    newMat = A+I
    oldMat = newMat
    flag = 0
    step = 1
    while flag == 0:
        oldMat = newMat
        newMat = oldMat*(A+I)
        for i in range(len(newMat)):
            for j in range(len(newMat)):
                if newMat[i, j] >= 1:
                    newMat[i, j] = 1
        step += 1
        #print(step)
        if (oldMat == newMat).all():
            flag = 1
            return newMat#返回可达矩阵
```

##算法二kruskal

```python
def kruskal(matrix):
    sort_edge=sort_edges(matrix)#对所有的边按权重进行排序
    adj_matrix=np.zeros((len(data),len(data)))
    edge_kruskal=[]#初始化最小生成树边集为空
    j=0
    i=0
    #for i in range(len(data)-1):
       # flag=0
    while (i<(len(data)-1)):
        reach_matrix=get_reach_matrix(adj_matrix)#更新可达矩阵
        if(reach_matrix[sort_edge[j][0][0],sort_edge[j][0][1]]==0):
            edge_kruskal.append((sort_edge[j][0][0],sort_edge[j][0][1]))
            i=i+1
            adj_matrix[sort_edge[j][0][0]][sort_edge[j][0][1]]=1
            adj_matrix[sort_edge[j][0][1]][sort_edge[j][0][0]]=1
            flag=1
            j=j+1
        else:
            j=j+1
    return edge_kruskal
```

```python
edge_kruskal=kruskal(edge_matrix)
edge_kruskal
```

```python
map_temp.remove_edges_from(map_temp.edges())
temp=dict()
for i,j in edge_kruskal:
    display.clear_output(wait=True)
    citys_distance=distance.distance((pos[data.city[i]][1],pos[data.city[i]][0]),(pos[data.city[j]][1],pos[data.city[j]][0])).km
    edge_matrix[i][j]=citys_distance
    temp[data.city[i],data.city[j]]=int(citys_distance)
    map_temp.add_edge(data.city[i],data.city[j],weight=citys_distance)
    nx.draw(map_temp,pos=pos,node_color='y',edge_color='y',node_size=50,with_labels=True)
    nx.draw_networkx_edge_labels(map_temp,pos=pos,edge_labels=temp)
    plt.show()
    plt.pause(0.4)
```
